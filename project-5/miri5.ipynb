```python
import numpy as np

def s(x):
    return 1/(1+np.exp(-x))

X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([[0.85],[0],[0],[1]])

W_1 = np.random.randn(2,4)
W_22 = np.random.randn(4,1)
bb1 = np.random.randn(1,4)
B222 = np.random.randn(1,1)

for z in range(5000):
    h = s(X.dot(W_1)+bb1)
    outt = s(h.dot(W_22)+B222)

print("result = ", outt)

#ÙŽchat Ai
```
